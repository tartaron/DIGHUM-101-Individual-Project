# How GPT-3 Reads the Book, Anna Karenina

  I'd like to show you how GPT-3 reads a book. If I send the book content, then GPT-3 will return a response that it thinks is appropriate.  
  The contents of the book were not preprocessed to be easy for GPT-3. In other words, I tried to make the same conditions as people.  

```python
def reading(paperContent):
    tldr_tag = "\n tl;dr:"
    openai.organization = ''
    openai.api_key = ""
    engine_list = openai.Engine.list() 
    
    resultText = ''
    for page in paperContent:    
        text = page.extract_text() + tldr_tag
        response = openai.Completion.create(engine="davinci",prompt=text,temperature=0.3,
            max_tokens=140,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0,
            stop=["\n"]
        )
        print(response["choices"][0]["text"])
        resultText += str(response["choices"][0]["text"] + '\n')
    return resultText

paperFilePath = "Anna Karenina Origin.pdf"
paperContent = pdfplumber.open(paperFilePath).pages

tempSummary = reading(paperContent)

# generate text file 
ai_dir_path = os.listdir("./Data/individual_openAI/")
i = len(ai_dir_path) + 1
name = 'Data/individual_openAI/openAI_summary_' + str(i) + '.txt'
with codecs.open(name, "w", "utf-8-sig") as temp:
    temp.write(tempSummary)
```

GPT-3 reads the text in real time and returns the responses.

### Preprocessing: Person & AI


Each book report's content is preprocessed to find the frequency of words.

#### Person

```python
person_dir_path = os.listdir("./Data/individual_person/")

# Designate an empty dictionary to store the filename and text as columns
person_for_dataframe = {}

i=1
# Loop through the directory of text files and open and read them
for file in person_dir_path:
    with open("./Data/individual_person/" + file, "r", encoding="utf-8") as to_open:
         person_for_dataframe[file] = to_open.read()
         #result = model(person_for_dataframe[file], num_sentences=3)
         #full = ''.join(result)
         #print('\n\n -------------------------------' + str(i) + '------------------------------------- \n\n' + full)
    i += 1
```

```python
# Create and append the dataframe with two columns - the file name and the text itself
person_data = (pd.DataFrame.from_dict(person_for_dataframe, 
                                       orient = "index")
                .reset_index().rename(index = str, 
                                      columns = {"index": "File", 0: "Text"}))

person_data["Text_processed"] = person_data["Text"].apply(gensim.utils.simple_preprocess)
# Using gensim for preprocessing using .apply()
processed = person_data["Text"].apply(gensim.utils.simple_preprocess)

# Stopword removal using NLTK stopword list and a lambda function
stop = stopwords.words('english')
no_stop = processed.apply(lambda x: [w for w in x if w not in stopwords.words('english')]) 

# Convert list back to str
person_data["Text_processed"] = [' '.join(t) for t in no_stop]

# Get top-10 words
person_data_str = ' '.join(person_data['Text_processed'].tolist())
person_data_tok = person_data_str.split()
person_data_freq = Counter(person_data_tok)

# Print the 10 most common words
person_data_df = pd.DataFrame(person_data_freq.most_common(10), columns = ["Person Word", "Frequency"])

# Save to csv!
person_data.to_csv('./Data/individual_person.csv', index=False)

# Define an empty bag (of words)
vectorizer = CountVectorizer()

# Use the .fit method to tokenize the text and learn the vocabulary
vectorizer.fit(person_data["Text_processed"])

# Encode the documents
vector = vectorizer.transform(person_data["Text_processed"])

# Note we are entering regular expression as a token_pattern argument
bigram_vectorizer = CountVectorizer(ngram_range = (1,2),
                                    stop_words = "english",
                                    token_pattern = r'\b\w+\b', 
                                    min_df = 1)

# Analyze string in the bigram bag of words
analyze = bigram_vectorizer.build_analyzer()
person_vocab = analyze(person_data_str)
```

#### AI

```python
ai_dir_path = os.listdir("./Data/individual_openAI/")

# Designate an empty dictionary to store the filename and text as columns
ai_for_dataframe = {}

i=1
# Loop through the directory of text files and open and read them
for file in ai_dir_path:
    with open("./Data/individual_openAI/" + file, "r", encoding="utf-8") as to_open:
         ai_for_dataframe[file] = to_open.read()
         #result = model(ai_for_dataframe[file], num_sentences=3)
         #full = ''.join(result)
         #print('\n\n -------------------------------' + str(i) + '------------------------------------- \n\n' + full)
    i += 1
```

```python
# Create and append the dataframe with two columns - the file name and the text itself
ai_data = (pd.DataFrame.from_dict(ai_for_dataframe, 
                                       orient = "index")
                .reset_index().rename(index = str, 
                                      columns = {"index": "File", 0: "Text"}))

ai_data["Text_processed"] = ai_data["Text"].apply(gensim.utils.simple_preprocess)

# Using gensim for preprocessing using .apply()
processed = ai_data["Text"].apply(gensim.utils.simple_preprocess)

# Stopword removal using NLTK stopword list and a lambda function
stop = stopwords.words('english')
no_stop = processed.apply(lambda x: [w for w in x if w not in stopwords.words('english')]) 

# Convert list back to str
ai_data["Text_processed"] = [' '.join(t) for t in no_stop]

# Get top-10 words
ai_data_str = ' '.join(ai_data['Text_processed'].tolist())
ai_data_tok = ai_data_str.split()
ai_data_freq = Counter(ai_data_tok)

# Print the 10 most common words
ai_data_df = pd.DataFrame(ai_data_freq.most_common(10), columns = ["AI Word", "Frequency"])

# Save to csv!
ai_data.to_csv('./Data/individual_AI.csv', index=False)

# Define an empty bag (of words)
vectorizer = CountVectorizer()

# Use the .fit method to tokenize the text and learn the vocabulary
vectorizer.fit(ai_data["Text_processed"])

# Encode the documents
vector = vectorizer.transform(ai_data["Text_processed"]) 

# Note we are entering regular expression as a token_pattern argument
bigram_vectorizer = CountVectorizer(ngram_range = (1,2),
                                    stop_words = "english",
                                    token_pattern = r'\b\w+\b', 
                                    min_df = 1)

# Analyze string in the bigram bag of words
analyze = bigram_vectorizer.build_analyzer()
ai_vocab = analyze(ai_data_str)
```

# BERT Extractive Summarization

In the case of extractive summarization generated by the BERT model, sentences determined to be important among the input texts are extracted as they are. In other words, the BERT model does not directly generate sentences.

In order to examine BERT extractive summarization, I decided that there was no need to compare AI's and people's writings, so I do summarize with only people's writings.

```python
#generate SBert model to summarize the text
model = SBertSummarizer('paraphrase-MiniLM-L6-v2')
```
